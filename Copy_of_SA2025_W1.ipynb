### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
#your code here
median_mpg = df['mpg'].median()
print("Median MPG:", median_mpg)
df['hp_to_weight_ratio'] = df['horsepower'] / df['weight']
filtered_cars = df[df['mpg'] > median_mpg].copy()
sorted_filtered_cars = filtered_cars.sort_values(by='hp_to_weight_ratio', ascending=False)
print("\nCars with above-median mpg, sorted by hp/weight ratio (descending):")
print(sorted_filtered_cars['name'])
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
#your code here
average_mpg_by_year_origin = df.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=average_mpg_by_year_origin, x='model_year', y='mpg', hue='origin')
plt.title('Evolution of Average MPG Over the Years by Origin')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.grid(True)
plt.show()
```

### We define a “consistent” car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
#your code here
# Identify car names that appear in more than one model_year and compute std dev of mpg
mpg_stats_by_name = df.groupby('name').agg(
    appearances=('model_year', 'nunique'),
    std_mpg=('mpg', 'std'),
    avg_mpg=('mpg', 'mean')
).reset_index()

# Filter for cars with at least 2 appearances and std(mpg) < 1.0
consistent_cars = mpg_stats_by_name[
    (mpg_stats_by_name['appearances'] >= 2) &
    (mpg_stats_by_name['std_mpg'] < 1.0)
]

# Sort the result by number of appearances (descending), then mpg (descending)
sorted_consistent_cars = consistent_cars.sort_values(by=['appearances', 'avg_mpg'], ascending=[False, False])

# Report the model name(s), number of appearances, and the average mpg
print("Consistent Car Models (std(mpg) < 1.0, >= 2 appearances):")
display(sorted_consistent_cars[['name', 'appearances', 'avg_mpg']])
```
